# Task: plotting a White Noise process

## Version 1

```python
import numpy
import matplotlib.pyplot as plt

ϵ_values = numpy.random.randn(100)
plt.plot(ϵ_values)

plt.show()
ϵ_values
```

## Implementing with a for loop

```python
ts_length = 100
ϵ_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    ϵ_values.append(e)

plt.plot(ϵ_values)
plt.show()

ϵ_values
```

## Implementing with a while loop

```python
ts_length = 100
ϵ_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
plt.plot(ϵ_values)
plt.show()

ϵ_values
```

## Another application
Plotting the balance of a bank account that yields interest.

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```
