```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Exercise 1
Given function: $f(x)=\cos(\pi\theta x)\exp(-x)$

```python
def givenFunction(x, theta):
    return np.cos(np.pi * theta * x) * np.exp(-x)

thetaArray = np.linspace(0, 2, 10)
xArray = np.linspace(0, 5, 200)

fig, ax = plt.subplots()

for value in thetaArray:
    currentLabel = f'$\Theta$ = {value}'
    ax.plot(xArray, givenFunction(xArray, value), label = currentLabel)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title('$f(x)=\cos(\pi \Theta x)\exp(-x)$')    
plt.show()
```
