```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 1

```python
x_t = np.empty(T+1)
x_t[0] = 0

T = 200
alpha = 0.9
iterator = 0

while(iterator < T):
    x_t[iterator+1] = alpha * x_t[iterator] + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Simulated time series')
plt.legend()
plt.show()
```

## Exercise 2

```python
x_t = np.empty(T+1)
x_t[0] = 0
T = 200

# Creates the first time series
alpha = 0
iterator = 0

while(iterator < T):
    x_t[iterator+1] = alpha * x_t[iterator] + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Alpha = 0')
plt.legend()

# Creates the second time series
alpha = 0.8
iterator = 0

while(iterator < T):
    x_t[iterator+1] = alpha * x_t[iterator] + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Alpha = 0.8')
plt.legend()

# Creates the third time series
alpha = 0.98
iterator = 0

while(iterator < T):
    x_t[iterator+1] = alpha * x_t[iterator] + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Alpha = 0.98')
plt.legend()

plt.show()
```

## Exercise 3

```python
x_t = np.empty(T+1)
x_t[0] = 0

T = 200
alpha = 0.9
iterator = 0

while(iterator < T):
    x_t[iterator+1] = alpha * abs(x_t[iterator]) + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Simulated time series')
plt.legend()
plt.show()
```

## Exercise 4

```python
x_t = np.empty(T+1)
x_t[0] = 0

T = 200
alpha = 0.9
iterator = 0

while(iterator < T):
    if(x_t[iterator] < 0):
        x_t[iterator+1] = alpha * (-1) * x_t[iterator] + np.random.randn()
    else:
        x_t[iterator+1] = alpha * x_t[iterator] + np.random.randn()
    iterator = iterator + 1
    
plt.plot(x_t, label='Simulated time series')
plt.legend()
plt.show()
```

## Exercise 5

```python
n = 1000000

count = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        count += 1

area_estimate = count / n * 4

print(area_estimate)  # dividing by radius**2
```
